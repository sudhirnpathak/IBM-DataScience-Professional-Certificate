# Recommendation Systems

There are similarities in the things people tend to like. In other words, people tend to like things in the same category, or things that share the same characteristics. For example, if some one has bought a book 'Machine learning with Python' and liked it, it is likely that the person will also buy 'Data Visualization with Python'. 

Recommender systems capture the pattern of people's behaviour and use it to predict what else they might want or like. These systems have become ubiquitous, and can be commonly seen in online stores, movies databases and job finders. There are generally two main types of recommendation systems :

 - Content based : Content based systems figure out what are the user's favorite aspects of an item are, and then make recommendation on items that share those aspects.
 - Collaborative filtering : Collaborative filtering techniques find similar group of users and provide recommendations based on similar taste within that group

# Content Based Recommendation Systems

A content based recommendation system tries to recommend items to user based on their profile. The user's profile revolves around that user's preferences and tastes. It is shaped based on user's ratings including the number of times the user has clicked on different items or perhaps even liked those items. The recommendation process is based on the similarity between those items. For example, if the user is watching and liking a comedy movie, the algorithm will suggest other comedy movies.

In this notebook, we will explore Content-based recommendation systems

Assume that we have a dataset of 5 movies (M1, M2, M3, M4, M5) with the details about the genre of each movie. Let's say that the user has watched 3 movies (M1, M2, M3) and given ratings (R1, R2, R3) (say some number out of 10) to these movies. The task of the recommender engine is to recommend one of the two candidate movies (M4 or M5) to this user. In other words, we want to predict what the user's possible rating would be of the two candidate movies, if the user were to watch them. With this prediction, we can suggest the movie with higher rating to the user.

**Building the user profile (See image_cbrs1.jpg)** : To achieve this, we have to build the user profile : First, we create a vector to show the user's rating for the movies that the user has already  watched, R = [R1, R2, R3] (visualize this vector vertically). We call this input user ratings. Then, we encode the movies into a matrix M using one-hot encoding approach. The movie matrix rows will be the movies which the user has rated and in the same order as the input user rating vector. The number of columns in the matrix will be the number of genres (features) used to characterize the movies. For example, five columns, if we characterize the movies using genres comedy, drama, adventure, romance, super-hero. Note that a movie can have more than one genre, like romantic comedy. Thus, with the user having watched 3 movies, and the movies having 5 genres, our movie matrix is 3 into 5. As in one-hot encoding, entry for a movie in a particular genre column will be 1 if the movie has that genre otherwise 0.

Multiplying each row of M with the scalar in the corresponding row of R, we get the weighted genre matrix (in general, weighted feature matrix). In other words, all elements in the row corresponding to the movie M1 in M, are multiplied by R1, while those in row corresponding to M2 are multiplied by R2. Now, that we have the weighted genre matrix, we can sum across the rows, which gives the cumulative sum for features (genres). We then normalize the cumulative value of each feature by the sum total of cumulative values of all features. Now, we have the user profile, where each genre has a value between 0 and 1, giving the fractional value of a particular genre for this particular user. The largest fraction genre is the favorite genre of this user and we recommend movies of this genre.

**Finding the recommendation (See image_cbrs2.jpg)** : Now that we know user's preferance, how do we recommended to him movies which he has not yet watched. In other words, how do we judge which movie to suggest M4 or M5. To this end, we can again prepare the one-hot encoding matrix for these two movies, with 1 in columns where the movie has that particular genre otherwise zero. Then we multiply each column (feature) by the fractional value user has for this particular feature. We then sum across all features to obtained the total value obtained for a particular movie. In other words, we are weighing all features of a movie according to the weight user has for these features, and then summing up. Thus, for each movie we have a value between 0 and 1 in our recommendation matrix. We can sort this and suggest the movie with highest value.

```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
movies_df = pd.read_csv('movies.csv')                  # movies information
ratings_df = pd.read_csv('ratings.csv')                # user information
```

```python
movies_df.head()
```

```python
ratings_df.shape
```

Let's remove the year from the __title__ column, and store in a new __year__ column.

```python
# Using regular expressions to find a year stored between parentheses
# We specify the parantheses so we don't conflict with movies that have years in their titles
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))', expand = False)

# Removing the parentheses
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand = False)

# Removing the years from the 'title' column
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '')

#Applying the strip function to get rid of any ending whitespace characters that may have appeared
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
```

```python
movies_df.head()
```

Now, let's also split the values in the __Genres__ column into a __list of Genres__ to simplify future use:

```python
# Every genre is separated by a | so we simply have to call the split function on |
movies_df['genres'] = movies_df.genres.str.split('|')
```

```python
movies_df.head()
```

Since keeping genres in a list format isn't optimal for the content-based recommendation system technique, we will use the one-hot encoding technique to convert the list of genres to a vector where each column corresponds to one possible value of the feature. This encoding is needed for feeding categorical data. In this case, we store every different genre in columns that contain either 1 or 0. 1 shows that a movie has that genre and 0 shows that it doesn't. For clarity, let's also store this dataframe in another variable, and we will use this only when we need to have information about genres.

```python
# Copying the movie dataframe into a new one
moviesWithGenres_df = movies_df.copy()

# For every row in the dataframe, iterate through the list of genres and place a 1 into the corresponding column
for index, row in movies_df.iterrows():
    for genre in row['genres']:
        moviesWithGenres_df.at[index, genre] = 1

#Filling in the NaN values with 0 to show that a movie doesn't have that column's genre
moviesWithGenres_df = moviesWithGenres_df.fillna(0)
```

```python
moviesWithGenres_df.head()
```

Next, let's look at the ratings dataframe

```python
ratings_df.head()
```

Every row in the ratings dataframe has a user id associated with at least one movie, a rating and a timestamp showing when they reviewed it. We won't be needing the timestamp column, so let's drop it to save memory.

```python
# Drop removes a specified row or column from a dataframe
ratings_df = ratings_df.drop('timestamp', 1)
```

```python
ratings_df.head()
```

Now, let's implement **Content-Based Recommendation Systems**. This technique attempts to figure out what a user's favourite aspects of an item are, and then recommends items that present those aspects. In our case, we're going to try to figure out the input's favorite genres from the movies and ratings given.

Let's begin by creating an input user. We assume that our input user has watched these movies and rated them as per his likes.

```python
userInput = [
    {'title':'Breakfast Club, The', 'rating':5},
    {'title':'Toy Story', 'rating':3.5},
    {'title':'Jumanji', 'rating':2},
    {'title':'Pulp Fiction', 'rating':5},
    {'title':'Akira', 'rating':4.5}
]

inputMovies = pd.DataFrame(userInput)
```

```python
inputMovies
```

Now we need to know the genre of the movies that our input user has watched and rated. To access this information, we extract the input movie's ID's from the movies dataframe and add them into it. 

We can achieve this by first filtering out the rows that contain the input movie's title and then merging this subset with the input dataframe. We also drop unnecessary columns for the input to save memory space.

```python
# Filtering out the movies by title
inputId = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]

# Then merging it so we can get the movieId. It's implicitly merging it by title.
inputMovies = pd.merge(inputId, inputMovies)

# Dropping information we won't use from the input dataframe
inputMovies = inputMovies.drop('genres', 1).drop('year', 1)
```

```python
# Final input dataframe
inputMovies
```

Now that we know the movieId of our input user, we can get all the details of the movies that our input user has watched.

```python
# Filtering out the movies from the input
userMovies = moviesWithGenres_df[moviesWithGenres_df['movieId'].isin(inputMovies['movieId'].tolist())]
```

```python
userMovies
```

We'll only need the one-hot encoded genre table, so let's clean this up a bit by resetting the index and dropping the movieId, title, genres and year columns.

```python
# Resetting the index to avoid future issues
userMovies = userMovies.reset_index(drop=True)

# Dropping unnecessary columns
userGenreTable = userMovies.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
```

```python
userGenreTable
```

Now, to obtain the weighted genre matrix, we multiply input's genre table by the input's ratings, and then summing up the resulting table by column. This operation is actually a dot product between a matrix and a vector, so we can simply accomplish by calling Pandas's 'dot' function.

```python
inputMovies['rating']
```

```python
# Dot produt to get weights
userProfile = userGenreTable.transpose().dot(inputMovies['rating'])

# The user profile
userProfile
```

Now, we have the user profile, i.e., weights for every of the user's preferences. We can now recommend movies that satisfy the user's preferences.

Let's start by extracting the genre table from the original dataframe:

```python
# Now let's get the genres of every movie in our original dataframe
genreTable = moviesWithGenres_df.set_index(moviesWithGenres_df['movieId'])

# And drop the unnecessary information
genreTable = genreTable.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
```

```python
genreTable.head()
```

```python
genreTable.shape
```

With the input's profile and the complete list of movies and their genres in hand, we're going to take the weighted average of every movie based on the input profile and recommend the top twenty movies that most satisfy it.

```python
# Multiply the genres by the weights and then take the weighted average
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df.head()
```

```python
# Sort our recommendations in descending order
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)

# Just a peek at the values
recommendationTable_df.head()
```

```python
# The final recommendation table
movies_df.loc[movies_df['movieId'].isin(recommendationTable_df.head(20).keys())]
```

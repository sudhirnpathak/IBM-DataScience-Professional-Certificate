# Decision Trees

Decision tree is a popular machine learning classification algorithm. Decision trees are built by splitting the training set into distinct nodes. Nodes are feature variables, which split the training set into number of categories that particular feature node has. For example, if the feature is age, and it has three categories (young, middle, senior), it will split the datasets into three branches. One can further divide these branches depending on other features like gender etc.

A decision tree is constructed by considering the attributes one by one. The algorithm is :

 - split data based on the value of the best attribute
 - then go to each branch, and repeat the above step for the rest of attributes

```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
```

Suppose we have historical data about a set of patients, all of whom suffered from the same illness. During their course of treatment, each patient responded to one of 5 medications, drugA, drugB, drugC, drugX and drugY. We want to build a decision tree model to find out which drug might be appropriate for a future patient with the same illness. 

The feature sets of this dataset are Age, Sex, Blood Pressure, and Cholesterol of patients, and the target is the drug that each patient responded to.

## Data Pre-processing

```python
df = pd.read_csv('drug200.csv')
```

Sex, BP and Cholesterol are categorical variables. Let's convert them into dummyVariables

```python
dummyVariable_sex = pd.get_dummies(df['Sex'])              # get dummy variables for the sex column
df = pd.concat([df, dummyVariable_sex], axis=1)            # concatenate the dummy variables to the dataframe
df.drop('Sex', axis = 1, inplace = True)                   # drop the sex column
df.drop('F', axis = 1, inplace = True)                     # drop the F sex dummy variable
df.rename(columns = {'M' : 'dSex'}, inplace = True)        # Male column will be sex vector : 0 for F and 1 for M

dummyVariable_chl = pd.get_dummies(df['Cholesterol'])
df = pd.concat([df, dummyVariable_chl], axis=1)
df.drop('Cholesterol', axis = 1, inplace = True)
df.drop('HIGH', axis = 1, inplace = True)
df.rename(columns = {'NORMAL' : 'dCholesterol'}, inplace = True) # Normal column will be cholesterol vector

dummyVariable_bp = pd.get_dummies(df['BP'])
df = pd.concat([df, dummyVariable_bp], axis=1)
df.drop('BP', axis = 1, inplace = True)
df['dBP'] = 2*df['NORMAL'] + df['LOW']              # 0,1,2 represents HIGH, LOW, NORMAL
df.drop('HIGH', axis = 1, inplace = True)
df.drop('LOW', axis = 1, inplace = True)
df.drop('NORMAL', axis = 1, inplace = True)
```

```python
df.head()
```

Let's define the feature matrix X and the target vector y:

```python
X = df[['Age', 'dSex', 'dBP', 'dCholesterol', 'Na_to_K']].values
y = df['Drug']
```

Let's split the data into train-test datasets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 3)
```

## Modeling (Decision Tree)

### How to build the decision tree ?

Decision trees are built using recursive partitioning to classify the data. The algorithm chooses the most predictive feature to split the data on. What is important in making a decision tree, is to determine which attribute is the best or more predictive to split the data.

Let's first define Entropy and Information gain, in order to quantify and determine best attribute to split the data

**Entropy** is a measure of randomness. The entropy in a node depends on how much random data is there in that node, and is given by:

<br>
$$\text{Entropy} E = -\sum p_i\log{p_i}$$
<br>
where $i$ represents a class, and $p_i$ is the fraction of data points in class $i$. 

For example, if we have 20 samples, and only two class, A and B. Assume, 12 samples are in class A and the remaining 8 are in class B. The entropy of the node is given by E = -(12/20)log(12/20) - (8/20)log(8/20). If all elements in a node belong to same target class, its entropy will be zero (called pure node). It will be greater than zero for any other case.

**Information gain** is the information that can increase the level of certaintly after splitting :

<br>
$$\text{Information gain = Entropy before split - weighted entropy after split}$$

For example, suppose we have $N$ number of samples and we know their classes. We can easily compute the entropy before split $S_B$. Suppose we divide the tree using attribute Sex. Their will be two branches, Male and Female. Say there are $N_M$ samples in male group and $N_F$ in female group. We can now calculate the entropy of these two groups, namely $S_M$ and $S_F$. The weighted entropy after dividing using Sex attribute is given by 

$$S_W = \frac{N_M}{N}S_M + \frac{N_F}{N}S_F$$
Now, we can easily calculate the information gain as $S_B-S_W$.

The best attribute is the one for which **information gain is maximum.** In other words, the best attribute is the one which decreases entropy the most in the resulting branches, and hence is best in making predictions, as this attribute leads to purer branches as compared to other attributes.

Let's understand this by a hypothetical example. Suppose we have only two classes of drugs, drugM and drugF. Assume that in this hypothetical example, all males responded to drugM and all females responded to drugF. Now, if we use sex attribute to make the tree, it will give two pure brances, one where it will be all males responding to drugM, and other where all females responding to drugF. The entropy of both these groups will be zero, giving the weighted entropy to be zero. This will give the maximum value of information gain possible, as entropy before split will be same for all the attributes. It is clear from this example, how highest information gaining attribute will result in more purer branches.

Let's import DecisionTreeClassifier, and create an instance of it named drugTree

```python
from sklearn.tree import DecisionTreeClassifier

# inside of the classifier, we specify criterion = 'Entropy' so we can see the information gain of each node
drugTree = DecisionTreeClassifier(criterion = 'entropy', max_depth = 4)
drugTree   # shows the default parameters
```

Now, we will fit the data with the training feature matrix X_train, and target data y_train

```python
drugTree.fit(X_train, y_train)
```

Let's make prediction on the test dataset and store it in the variable called predTree

```python
predTree = drugTree.predict(X_test)
```

Let's compare the prediction on the test set with the actual value of the target variable

```python
print(y_test[0:5])
```

```python
print(predTree[0:5])
```

Now, let's estimate the accuracy of our model

```python
from sklearn import metrics

print('Accuracy: ', metrics.accuracy_score(y_test, predTree))     # fraction of correctly classified samples
```
